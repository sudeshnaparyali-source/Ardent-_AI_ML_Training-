**Loops**: for, while

```python
print("print 1-10 using for loop")
# print 1-10 using for loop
for i in range(1,11):
  print(i)

print("print even number from 1-10")
# print even number from 1-10
for i in range(2,11,2):
  print(i)

print("print odd number from 3-20")
# print odd number from 3-20
for i in range(3,21,2):
  print(i)

print("print reverse number (10-1)")
# print reverse number (10-1)
for i in range(10,0,-1):
  print(i)


print("Try to understand Nested Loops")
# Try to understand Nested Loops & Importance of Indentation
"""
for i in range(10,3,-2):
  print(i)
  for i in range(10,3,-2):
    print(i)
    for i in range(10,3,-2):
      print(i)
"""
```
